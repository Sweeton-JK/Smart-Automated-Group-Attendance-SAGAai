```python
from facenet_pytorch import MTCNN
from PIL import Image
import os
```

```python
output_folder = '../Images/output_images/testImages1/'
os.makedirs(output_folder, exist_ok=True)
```

```python
mtcnn = MTCNN(
    image_size=240,
    margin=20,
    min_face_size=20,
    thresholds=[0.6, 0.7, 0.7], 
    factor=0.709,  
    post_process=True, 
)
```

```python
img = Image.open('./oscar1.jpg')
```

```python
boxes, probs = mtcnn.detect(img)
```

```python
if boxes is not None:
    # Create a folder to store the cropped faces
    output_folder = '../Images/output_images/testImages1/'
    os.makedirs(output_folder, exist_ok=True)

    # Define the expansion factor (adjust as needed)
    expansion_factor = 0.05

    # Iterate over detected faces
    for i, box in enumerate(boxes):
        # Expand the bounding box
        expanded_box = [
            box[0] - int(expansion_factor * box[2]),  # x1
            box[1] - int(expansion_factor * box[3]),  # y1
            box[2] + int(expansion_factor * box[2]),  # x2
            box[3] + int(expansion_factor * box[3])   # y2
        ]

        # Crop the face
        face_img = img.crop(expanded_box)

        # Save the cropped face image
        face_img_path = os.path.join(output_folder, f'face_{i+1}.png')
        face_img.save(face_img_path)
        print(f"Face {i+1} saved at: {face_img_path}")

    print("Faces detected and saved.")
else:
    print("No faces detected.")
```
